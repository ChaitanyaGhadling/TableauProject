```python
import pandas as pd
import numpy as np
```

```python
shooting_data = pd.read_csv('NYPD_Shooting_Incident_Data__Historic_.csv')
```

```python
daily_inmates_data = pd.read_csv('Daily_Inmates_In_Custody.csv')
```

```python
arrest_data = pd.read_csv('NYPD_Arrests_Data__Historic_.csv')
```

```python
shooting_data['PERP_RACE'] = shooting_data['PERP_RACE'].map({'BLACK': 'B', 'WHITE': 'W','UNKNOWN':'U','ASIAN / PACIFIC ISLANDER':'A','AMERICAN INDIAN/ALASKAN NATIVE':'I', 'WHITE HISPANIC':'O','BLACK HISPANIC':'O'})
arrest_data['PERP_RACE'] = arrest_data['PERP_RACE'].map({'BLACK': 'B', 'WHITE': 'W','UNKNOWN':'U','ASIAN / PACIFIC ISLANDER':'A','AMERICAN INDIAN/ALASKAN NATIVE':'I', 'WHITE HISPANIC':'O','BLACK HISPANIC':'O'})
```

```python
print(shooting_data.PERP_RACE.value_counts())
print("")
print(daily_inmates_data.RACE.value_counts())
print("")
print(arrest_data.PERP_RACE.value_counts())
```

```python
print(shooting_data.PERP_SEX.value_counts())
print("")
print(daily_inmates_data.GENDER.value_counts())
print("")
print(arrest_data.PERP_SEX.value_counts())
```

```python
shooting_data = shooting_data[shooting_data['PERP_AGE_GROUP'].notna()]
```

```python
shooting_data = shooting_data.loc[shooting_data['PERP_AGE_GROUP'].isin(['18-24','25-44','<18','45-64','65+','UNKNOWN'])]
arrest_data = arrest_data.loc[arrest_data['AGE_GROUP'].isin(['18-24','25-44','<18','45-64','65+','UNKNOWN'])]
```

```python
bins = [0, 18, 24, 44, 64, np.inf]
names = ['<18', '18-24', '25-44', '45-64', '65+']
daily_inmates_data['AGE'] = pd.cut(daily_inmates_data['AGE'], bins, labels=names)
```

```python
print(shooting_data.PERP_AGE_GROUP.value_counts())
print("")
print(daily_inmates_data.AGE.value_counts())
print("")
print(arrest_data.AGE_GROUP.value_counts())
```

```python
daily_inmates_data[['Date','Time','A']] = daily_inmates_data.ADMITTED_DT.str.split(" ", expand=True)
daily_inmates_data.drop(columns=['A','ADMITTED_DT','Time'], inplace = True)
```

```python
shooting_data['OCCUR_DATE'] = pd.to_datetime(shooting_data['OCCUR_DATE'], format='%m/%d/%Y')
arrest_data['ARREST_DATE'] = pd.to_datetime(arrest_data['ARREST_DATE'], format='%m/%d/%Y')
daily_inmates_data['Date'] = pd.to_datetime(daily_inmates_data['Date'], format='%m/%d/%Y')
```

```python
shooting_data['Arrest_limit'] = pd.DatetimeIndex(shooting_data.OCCUR_DATE) + pd.DateOffset(3)
```

```python
df_merged = pd.merge(shooting_data, arrest_data, left_on=
                     ['PERP_SEX', 'PERP_RACE','PERP_AGE_GROUP','BORO','PRECINCT'], 
                     right_on = ['PERP_SEX','PERP_RACE','AGE_GROUP','ARREST_BORO','ARREST_PRECINCT'], 
                     how='inner')
```

```python
print(shooting_data.shape)
print(arrest_data.shape)
print(df_merged.shape)
```

```python
mask = (df_merged['ARREST_DATE'] >= df_merged['OCCUR_DATE']) & (df_merged['ARREST_DATE'] < df_merged['Arrest_limit'])
```

```python
print(mask.value_counts())
type(mask[0])
```

```python
df_merged = df_merged[mask]
```

```python
print(shooting_data.shape)
print(arrest_data.shape)
print(df_merged.shape)
```

```python
df_merged_again = pd.merge(df_merged, daily_inmates_data, left_on=
                           ['PERP_SEX', 'PERP_RACE','PERP_AGE_GROUP','OCCUR_DATE'], 
                           right_on = ['GENDER','RACE','AGE','Date'], how='inner')
```

```python
print(df_merged.shape)
print(df_merged_again.shape)
print(df_merged_again.INCIDENT_KEY.nunique())
print(df_merged_again.INMATEID.nunique())
```
